## REGISTRATION 

```python
file = open("login.txt", "a")

import re

print("**********     REGISTRATION FORM     **********")

#####################    EMAIL VERIFICATION ################################
email_verify = '^[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}$'

email = input("Enter your email id: ")
val_email= True
if (re.search(email_verify, email)):
    #     print("valid Email")

    with open('Login.txt', 'r') as e:
        for line in e:
            loginInfo = line.strip().split(';')
            if email == loginInfo[0]:
                val_email = False
                print('Error: that username is already in use.')
                print("try some other username")

                break
            else:
                val_email = True

else:
    print("Invalid Email")
    print("try again")
    val_email=False

################### PASSWORD VALIDATION ##########################

valid_pwd = True
if val_email:
    new_pwd = input("Enter Password :")
    
    while True:
        if (len(new_pwd) > 16 and len(new_pwd) < 5):
            valid_pwd = False
            break
        elif not re.search("[a-z]", new_pwd):
            valid_pwd = False
            break
        elif not re.search("[A-Z]", new_pwd):
            valid_pwd = False
            break
        elif not re.search("[0-9]", new_pwd):
            valid_pwd = False
            break
        elif not re.search("[_@$]", new_pwd):
            valid_pwd = False
            break
        else:

################# CONFIRM PASSWORD #########################
            confirm_pwd = input("confirm your password: ")
            if confirm_pwd == new_pwd:
                print("account created")
                file= open("login.txt","a")
                data= email + ";" + new_pwd
                file.write(data)
                file.write("\n")
                file.close()
                print("Your login details have been saved")
                break
            else:
                print("password did not match.Try again")
                break


    if valid_pwd == False:
        print("not a Valid Password")

```
